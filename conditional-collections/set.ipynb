```python
#1 Append the element in the set 
team=["RCB","MI","CSK","GT","LSG","RR","KKR","PNB","SRH"]
print(team)

team.append("DC") # add Dc into list team
print(team)
```

```python
#2 Add the element in the set.
team={"RCB","SRH","GT","LSG"}

# add the RR into team
team.add("RR")
print(team)
```

```python
#3.update() :- is used to add thet multiple items from other collections
team={"RCB","SRH","GT","LSG"}
player={"Virat","Pat","Miller","Samad"}
# add the items team into player
team.update(player)
print(team)
```

```python
# remove the items from the set
team={"RCB","SRH","GT","LSG"} 
# remove the GT form team
team.pop()
print(team)
#POP() is bad practice in sets because set is unordered


# remove(item_name)
team.remove("SRH")
print(team)
```

```python
#1. remove(item_name)
team={"RCB","SRH","GT","LSG"} 
team.remove("SRH")
print(team)
```

```python
# 2.discard(item_name)
# It is used to remove an element from a set.
bowler={"Shami","Bolt","Boomrah","Chahal","Stoke","Archer","Rabada"}
bowler.discard(2) i
print(bowler)
```

```python
#what is the difference between remove and discard?
#if the item doesnot exist in remove() it rises an error
#where as in discard() it wont rises any error even item exist or not

#remove --> Print the KeyError
bowler={"Shami","Bolt","Boomrah","Chahal","Stoke","Archer","Rabada"}
bowler.remove("medha")
print(bowler)
```

```python
# discard 
bowler={"Shami","Bolt","Boomrah","Chahal","Stoke","Archer","Rabada"}
bowler.discard("Medha")
print(bowler)
```
