```python
# List is Ordered
# **Positive indexing**
# The item in the list have sequence of index position numbers
# by grouping the index position numbers we can access the dat from list
# Print the pineapple from above list
# syntax (start ,end ,step)
l=["Apple","Banana","Pineapple","Graps","Avacado","Date"]
# print(list)
print(l[0:4]) # access the range of items from apple to graps

print(len(list)) #access the len of items.

print(type(list)) # Access the data type.

print(l[0:7:2])#access the even index position items

print(l[1:7:2]) #access the odd index items
```

```python
#Negative index --> access the range of elements in reverse direction
#Negative indexing starts from -1
l3 = ['apple', 'banana', 'pineapple', 'orange', 'grapes']
#print the last element from list l3
print(l3[-1])
print(l3[-1:-6:-1])
```

```python
#default indexing:-by default start is 0 and end is len(list) and step is 1.
n=["Banana","Pineapple","Graps","Avacado","Date"]
print(n[:])

#print the len and list.
print(n[0:len(n)])
# Even index character
print(n[::2])
#odd index character
print(n[1::2])
# Reverse the entire list
print(n[::-1])
# print the entire list
print(n[::])
```
