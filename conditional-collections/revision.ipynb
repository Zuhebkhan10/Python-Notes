```python
# **revision

# list --> [] --> ordered, mutable, and allows the duplicates
    # adding methods --> append,insert, extend
    # replace the items --> list[index]
    # remove methods --> pop,remove,clear,del

# Tuple --> () --> ordered, immutable and allows duplicates 
    #1. in tuple we cannot replace,add and remove the items bcz it is immutable

# SET ---> {}, set() --> unordered, immutable, doesn't allow duplicates
    # 1. set is immutable we cannot replace the data but we can add and remove the items from set
    # 2. adding methods of set --> add,update
    # 3. remove methods --> remove,pop,discard,clear

# # Dictionary --> {key:value} --> unordered, mutable, and wont allow duplicate values
    # 1. replace mehod of dictionary --> dict['existing_key_name']
    # 2. add methods of dictionary --> dict['New_key_name'],update 
    # 3. remove methods of dictionary --> pop("key"),popitems(),clear()
```
