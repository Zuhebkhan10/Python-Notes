```python
# Adding Elements into List
# We can add elements to a list using the following methods:

# append(): Adds an element at the end of the list.
# extend(): Adds multiple elements to the end of the list.
# insert(): Adds an element at a specific position.
# clear(): removes all items.
```

```python
# append(): Adds an element at the end of the list.
a=[10,20,40,50,90]
print(a)

a.append(100)
print("After the append value",a)
```

```python
# extend(): Adds multiple elements to the end of the list.
player=["Virat","Rajat","Salt","bethell","Hazelhood","Jistes"]
runs=["80","56","52","12","60","43"]
print(player)
print(runs)

player.extend(runs)
print(player)
```

```python
# insert(): Adds an element at a specific position.
runs=[10,20,55,98,33,12,17]

runs.insert(0,5)
runs.insert(8,90)
print(runs)
```

```python
# clear(): removes all items.
runs=[10,20,55,98,33,12,17]

runs.clear()
print(runs)
```

```python
# Removing Elements from List
# We can remove elements from a list using:

# remove(): Removes the first occurrence of an element.
# pop(): Removes the element at a specific index or the last element if no index is specified.
# del statement: Deletes an element at a specified index.
```

```python
# remove(): Removes the first occurrence of an element.
fav=["biryani","Dal rice","Green chicken","ColdCoffee","EggCurry","Tahri"]
print(fav)
print(len(fav))
fav.remove("Dal rice")
print("After remove the element is ",fav)
print(len(fav))
```

```python
# pop(): Removes the element at a specific index or the last element if no index is specified.
value=[45,53,80,75,102,115]
print(value)
value.pop() # pop the last element.
print("After the element is popped ",value)
```

```python
# del statement: Deletes an element at a specified index.
data=["Excel","PowerBI","SQL","Python"]
print(data)
del data[1] #Delete the item name is PowerBi
print("After delete the item is ",data)
```

```python
#Iterating Over list for loop and while loop 
fruits=["Apple","Pineaplle","Watermelon","Cherry","Graps"]
for i in fruits:
    print(i,end=" ")
```

```python
# Using while loops
fruits=["Apple","Pineaplle","Watermelon","Cherry","Graps"]
i=0
while i< len(fruits):
    print(fruits[i],end=" ")
    i=i+1
```

```python
# Nested Lists
m=[[1,2,3]
   ,[4,5,6]
   ,[7,8,9]]
print(m[0])
print(m[1])
print(m[2])

print("Nested list the matrix position is",m[1][2])
```

```python
# List Comprehension
# List comprehension is a concise way to create lists using a single line of code.
# It is useful for applying an operation or filter to items in an iterable, such as a list or range.

square=[x**2 for x in range(2,6)]
print("After entire element the sqrt ",square)

# Explanation::
# for x in range(1, 6): loops through each number from 1 to 5 (excluding 6).
# x**2: squares each number x.
# [ ]: collects all the squared numbers into a new list.
```
