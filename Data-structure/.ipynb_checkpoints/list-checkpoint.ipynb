```python
# List is Mutable

# >A collection mutable if it follows following 3 properties

# we can replace or swap the items
# we can add new items
# we can remove the items


```

```python
# Replace the item from the list
#in list we can replace,add and remove the items

a = "hello welcome"
#replace the welcome to goodmorning

print(a.replace("welcome","good morning"))
```

```python
l5 = ['apple', 'banana', 'pineapple', 'orange', 'grapes']
#replce 'orange', with watermelon
#we need to use index position number
l5[3] = "watermelon"
print(l5)
```

```python
#add the items from the list
print(dir(l5))
```

```python
#1.append():- by default it adds the item at end of the list
l = ['apple', 'banana', 'pineapple', 'watermelon', 'grapes']
# add the number 10 into l6
l.append(10)
print(l)
```

```python
#2.insert():- adds the item at specific index
#add the kiwi at 5th index in l6
l.insert(5,'kiwi')
print(l)
```

```python
#3.extend():- adds the multiple items from the other collection
l = [10,20,30,40]
#add the items of l7 into l6
l6.extend(l)
print(l6)
```

```python
l3 = ['apple', 'banana', 'pineapple', 'watermelon', 'grapes', 'kiwi', 10, 10, 20, 30, 40]
#1.pop():- by default removes the last item from the list
l3.pop()
print(l3)
```

```python
#2.remove(item)
l2 = ['apple', 'banana', 'pineapple', 'watermelon', 'grapes', 'kiwi', 10, 10, 20, 30, 40]
l2.remove("pineapple")
print(l2)
```

```python
#4.clear()
l3 = ['apple', 'banana', 'pineapple', 'watermelon', 'grapes', 'kiwi', 10, 10, 20, 30, 40]
l3.clear()
print(l3)
```

```python
#5.del
l=['apple', 'banana', 'pineapple', 'watermelon', 'grapes', 'kiwi']
del l
print(l)
```
