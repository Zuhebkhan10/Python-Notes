```python
#what is tuple ahd how can we declare a tuple?
#tuple is declred by using --> ()
#Tuple is defined as ordered, immutable and allows the duplicate data
```

```python
# Tuple allows the duplicates data

t=("Apple","Banana","Kiwi","Pineapple","Mango","Watermelon")
print(len(t)) #What is the length
print(t)
print(type(t)) #what is the type
```

```python
#Tuple is ordered --> The items is the tuple has sequence of index position numbers.
t=("Apple","Banana","Kiwi","Pineapple","Mango","Watermelon")
print(t[2]) # Print kiwi from above tuple

# print the items from banana kiwi pineapple mango
print(t[1:5])
```

```python
#print the even index psoition characters?
bowler=("Siraj","Dale","Pat cummins","BenStoke","Strac","Amir","Boomrah","Chahal")
print(bowler)
print(len(bowler))

print(bowler[0:8:2]) 

```

```python
# print the odd
batsman=("Virat","Deviliers","Jitesh","Salt","Brevis","Morgan","Joe root","Butler","Smith")
print(batsman[1:4:2])
```

```python
# Reverse the tuple
batsman=("Virat","Deviliers","Jitesh","Salt","Brevis","Morgan","Joe root","Butler","Smith")
print(batsman[::-1])
```
