```python
#dictionary is unordered
d={    
    "Name":"Medha",
    "Course":"DS/DA",
    "Duration":"4-6 Months",
    "Fee":"20000"
}
print(d)
print(len(d))
print(type(d))

# Key value pairss
print(d["Name"]) #Keys
print(d["Fee"]) #
```

```python
p={
    "game":"cricket",
    "batsman":"Virat",
    "bowler":"Hazelwood",
    "allrounder":"Laim Livingstong",
    "team":["RCB","RR","GT","LSG","KKR","CSK"]
}
# Access the "LSG" from above dictionary
print(p)
print(p["team"])
print(p["team"][3])

```

```python
# **Dictionary is Mutable**
# we can replace the data in dict
# we can add new data in dict
# we can remove the data in dict
```

```python
# replace the values from dictionary.
p={
    "game":"cricket",
    "batsman":"Virat",
    "bowler":"Hazelwood",
    "allrounder":"Laim Livingstong",
    "team":["RCB","RR","GT","LSG","KKR","CSK"]
}
# replace the batsman name to Phil salt.
p["batsman"]="Phil salt"
print(p)
```

```python
# add the new key-value pairs
p={
    "game":"cricket",
    "batsman":"Jacob bethell",
    "bowler":"Hazelwood",
    "allrounder":"Laim Livingstong",
    "team":["RCB","RR","GT","LSG","KKR","CSK"]
}
p["orange-cap"]="Virat"
print(p)
```

```python
#Pop 
d={
    "game":"cricket",
    "batsman":"Jacob bethell",
    "bowler":"Hazelwood",
    "allrounder":"Laim Livingstong",
    "team":["RCB","RR","GT","LSG","KKR","CSK"]
}
#remove the address from the "d"
#pop("key_name")
d.pop("team")
print(d)
```

```python
# remove the CSk
d={
    "game":"cricket",
    "batsman":"Jacob bethell",
    "bowler":"Hazelwood",
    "allrounder":"Laim Livingstong",
    "team":["RCB","RR","GT","LSG","KKR","CSK"]
}
print(d)
print(d["team"].remove("CSK"))
print(d)
```
