```python
# Logical Operator
#there are 3 types of logical operators
#AND:- if all the conditions are true  then it execute the code
#or:- if any one condition is true it executes the code
#not it reverse your actual o/p
print(10==10)
print(not(10==10))

#if we want to check the multiple conditions at a time we use the logical operatorsprint(not(10==10))
```

```python
# AND operator
#check that b is less then d and c is greater then a and d is greater then a
#if all the conditions are true print all conditions are satisfied
#else print not satisfied
a=10
b=45
c=55
d=90

if b<d and c>a and d>a:
    print("All condition are true")
else:
    print("Not statisfied")
```

```python
# Or
#check that a is greater then d and a is greater then b and c is greater then d
# if any one condition satisfied print condition satisfied
a=60
b=45
c=55
d=90

if a>d or a>b or c>d:
    print("All condition are true")
else:
    print("Not satisfied")
    
```

```python
# Ternary Operator
#if we want to check 2 conditions in a single line code we use the terinary operator
#write program to check eligible to vote or not
#if the age>=18 print eligible
#else print not eligible
age=(int(input("Enter your age ")))
if age>=18:
    print("Eligible")
else:
    print("Not Eligible")
    
```
