```python
# Write a program to check the student grades based on rescirved marks 
# If the student get marks between 33 to 45 print Congratulations you have recieved "D grade"
# If the student get marks between 46 to 60 print Congratulations you have recieved "C grade"
# If the student get marks between 61 to 75 print Congratulations you have recieved "B grade"
# If the student get marks between 76 to 85 print Congratulations you have recieved "A grade"
# If the student get marks between 86 to 100 print Congratulations you have recieved "A+ grade"

marks=int(input("Enter your number"))
if marks<=100:
    if marks>=33 and marks<=45:
        print("You have recieved D-grade")
    elif marks>=46 and marks<=60:
        print("You have recieved C-grade")
    elif marks>=61 and marks<=75:
        print("You have recieved B-grade")
    elif marks>=76 and marks<=85:
        print("Congratulations you have recieved A-grade")
    elif marks>=86 and marks<=100:
        print("Congratulations you have recieved A+ grade")
    else:
        print("Better luck next time")
```
