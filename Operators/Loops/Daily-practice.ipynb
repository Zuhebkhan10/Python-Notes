```python
#find the factorial of a number ex num=5 5*4*3*2*1
num=5
fac=1
for i in range(1,num+1):
    fac*=i
print(fac)    
```

```python
#Find the largest number in the list l=[10,21,34,55,66,78]
l=[10,21,34,55,66,78]
largest=l[0]
for num in l:
    if num>largest:
        largest=num
print(largest)
```
