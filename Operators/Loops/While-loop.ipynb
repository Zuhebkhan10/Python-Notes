While loop runs a set of instruction until the condition remains true.


```python
# Write a simple program to run a lify by using while loop
# if the user press 6th floor stop the lift at 6th floor.
user=int(input("Enter floor number"))
i=1
while(i<=user):
    print("You are at floor ",i)
    i=i+1
```

```python
# print 1 to 10 number by using the while loop
i=1
while i<=10:
    print(i,end=" ")
    i=i+1
```

```python
# Print the 10 to 1 number by using the Reverse order
a=10
while a>=1:
    print(a,end=" ")
    a=a-1
```

```python
#break the loop when it reches to 5 in the while loop
#Break Stop the iteration irrespective of next iterations
i=1
while i<=10:
    if (i==8):
        break
    print(i,end=" ")
    i=i+1
```
